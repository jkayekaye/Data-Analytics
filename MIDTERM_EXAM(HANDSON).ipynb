```python
#Imports
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import matplotlib as plt
```

```python
#Renaming of the data
chic_crimes= pd.read_csv('datasets\\Chicago_Crimes.csv')
```

```python
#Shows the dataset of Crimes in Chicago
chic_crimes
```

```python
#to show info about the datasets
chic_crimes.info()
```

```python
#Shows the Empty data
chic_crimes.isnull().sum()
```

```python
#description of data
chic_crimes.describe()
```

```python
#Clearing the null values or data
chic_crimes['Location Description']=chic_crimes['Location Description'].fillna('unaccounted')
chic_crimes['X Coordinate']=chic_crimes['X Coordinate'].fillna(chic_crimes['X Coordinate'].mode()[0])
chic_crimes['Y Coordinate']=chic_crimes['Y Coordinate'].fillna(chic_crimes['Y Coordinate'].mode()[0])
chic_crimes['Latitude']=chic_crimes['Latitude'].fillna(chic_crimes['Latitude'].mode()[0])
chic_crimes['Longitude']=chic_crimes['Longitude'].fillna(chic_crimes['Longitude'].mode()[0])
chic_crimes['Location']=chic_crimes['Location'].fillna(chic_crimes['Location'].mode()[0])
chic_crimes['Community Area']=chic_crimes['Community Area'].fillna(chic_crimes['Community Area'].mode()[0])


chic_crimes['Date'] = chic_crimes['Date'].astype(str)
chic_crimes['Date'] = chic_crimes['Date'].apply(lambda x: x.replace('/', '-') if pd.notnull(x) else x)
chic_crimes['Date'] = chic_crimes['Date'].apply(lambda x: x.strip() if pd.notnull(x) else x)
chic_crimes['Date'] = pd.to_datetime(chic_crimes['Date'], dayfirst=True, errors='coerce')


chic_crimes['Year'] = chic_crimes['Date'].dt.year
chic_crimes['Month'] =  chic_crimes['Date'].dt.month
chic_crimes['Day'] =  chic_crimes['Date'].dt.day
chic_crimes['DayOfWeek'] =  chic_crimes['Date'].dt.dayofweek
chic_crimes.isnull().sum()
```

```python
#Showing the data types
chic_crimes.dtypes
```

```python
#Changing into Category
chic_crimes['Case Number'] = chic_crimes['Case Number'].astype('category')
chic_crimes['Block'] = chic_crimes['Block'].astype('category')
chic_crimes['IUCR'] = chic_crimes['IUCR'].astype('category')
chic_crimes['Primary Type'] = chic_crimes['Primary Type'].astype('category')
chic_crimes['Description'] = chic_crimes['Description'].astype('category')
chic_crimes['Location Description'] = chic_crimes['Location Description'].astype('category')
chic_crimes['FBI Code'] = chic_crimes['FBI Code'].astype('category')
chic_crimes['Updated On'] = chic_crimes['Updated On'].astype('category')
chic_crimes['Location'] = chic_crimes['Location'].astype('category')
```
